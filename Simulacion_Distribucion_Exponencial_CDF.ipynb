# Objetivo:
 Simular una distribución exponencial usando el método de transformación inversa y comparar sus propiedades empíricas con los valores teóricos.

# Idea:
 Se generan variables aleatorias uniformes U ~ (0,1) y se transforman mediante:


# $X = -\frac{1}{\lambda} \ln(U)$

lo que produce variables con distribución exponencial.

# Qué se analiza:
 - Media y varianza empírica
 - Comparación con valores teóricos
 - Visualización mediante histograma y función de distribución acumulada (CDF)

# Aplicación:
 La distribución exponencial se usa en modelado de tiempos de espera, fallas de sistemas y procesos de Poisson (relevante en riesgo y eventos financieros)




```python
import math
import random as rnd
import numpy as np
import matplotlib.pyplot as plt



# Parámetros
λ=15            # parámetro de la distribución exponencial
n=10000         # número de simulaciones
L=-1/λ
num=100
numeros=[]

# Generación de datos
while num!=0:
  u=rnd.random()         # U ~ (0,1)
  x=L*np.log(u)          # transformación a exponencial
  numeros.append(x)
  num-=1

# Cálculo de estadísticas
media=np.mean(numeros)
varianza= np.var(numeros)

print("Media empírica:", media, "\n")
print("Media teórica (1/λ):", 1/λ, "\n")
print("Varianza empírica:", varianza, "\n")
print("Varianza teórica (1/λ^2):", 1/(λ**2), "\n")

# Histograma
plt.hist(numeros)


```

# Nota:
 A continuación se muestra la secuencia de valores generados en orden de simulación.

 Dado que se trata de variables aleatorias independientes, no se observa un patrón específico, lo cual es consistente con el comportamiento esperado.

```python
plt.plot(numeros)
```

```python
# Función de distribución acumulada empírica (CDF)
def cdf(numeros):
  n=len(numeros)
  x=np.sort(numeros)
  y=np.arange(0, len(numeros))/len(numeros)
  return x,y

X,Y=cdf(numeros)
plt.step(X,Y)
plt.step(X,Y, "cs", where="mid")
plt.xlim(0, 0.3)
plt.show()
```
